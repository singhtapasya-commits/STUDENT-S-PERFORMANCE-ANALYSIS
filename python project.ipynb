This project analyzes student performance using Python. The dataset contains student scores in math, reading, and writing along with factors like gender, lunch type, and test preparation.

The goal of this project is to understand patterns, relationships, and factors affecting student performance using data analysis and visualization.

## 📊 Distribution of Scores

Histogram shows how the data is distributed.

* Most students score between **60–80**
* Helps understand the overall performance trend


```python
sns.histplot(df['math score'], kde=True)
plt.title("Distribution of Math Scores")
plt.show()
```

Shows how math scores are distributed

Helps see:

~Most common score range

~Spread of data

~Most students score between 60–80

## 👥 Countplot

Countplot shows the number of students in each category.

Here, it shows the count of male and female students.


```python
sns.countplot(x='gender', data=df)
plt.title("Number of Students by Gender")
plt.show()
```

## 📦 Boxplot Analysis

Boxplot is used to compare distributions and detect outliers.

Here, we compare **gender vs math score**:

* It shows median, spread, and outliers
* Helps in comparing performance between male and female students


```python
sns.boxplot(x="gender", y="math score", data=df)
plt.show()
```

The boxplot shows that male students have a slightly higher median math score compared to female students. However, both groups have similar score distributions. Some outliers indicate very low-performing students.”

👉 “There is no huge difference, but a slight variation exists.”

## 🍽️ Lunch vs Performance

This analysis shows how lunch type affects student performance.

Students with standard lunch tend to perform better than those with free/reduced lunch.


```python
sns.boxplot(x="lunch", y="math score", data=df)
plt.show()
```

🔍 How to read the graph

1️⃣ Two boxes

~One for standard

~One for free/reduced

2️⃣ Middle line (median)

👉 This is the average-like middle score

~Standard → Higher 📈

~Free/reduced → Lower 📉

👉 Means:
👉 Students with standard lunch score better

3️⃣ Box size

👉 Shows spread of marks

👉 Both may have similar spread, but position differs

4️⃣ Dots (outliers)

👉 Very low or very high scores

👉 Shows some students perform very differently

🎯 Final Interpretation

“The boxplot shows that students with standard lunch have higher math scores compared to those with free or reduced lunch. This indicates that economic factors may influence academic performance.”

## 📊 Correlation Analysis

Correlation helps us understand the relationship between numerical variables.

* Values close to **1** indicate a strong positive relationship
* Values close to **0** indicate no relationship

We use a heatmap to visualize the correlation between math, reading, and writing scores.


```python
df.corr(numeric_only=True)
```

📊 How to read this

🔹 1. Diagonal = 1.0

👉 Same column with itself

👉 Always 1 (perfect relation)




🔹 2. Values close to 1

👉 Strong positive relation

Example:

Reading & Writing = 0.95 🔥

👉 Means:
👉 Students good in reading are also good in writing




🔹 3. Values around 0.8

👉 Still strong relation

Example:

Math & Reading = 0.82




🔹 4. If value was near 0

👉 No relation


🎯 Final Interpretation

The correlation matrix shows a strong positive relationship between reading and writing scores, and a moderate to strong relationship between math and other subjects.”

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()

```

## 📝 Test Preparation Analysis

Students who completed the test preparation course generally scored higher.

This indicates that preparation has a positive impact on performance.


```python
sns.boxplot(x='test preparation course', y='reading score', data=df)
plt.show()
```

📊 What this graph shows

👉 X-axis: Test Preparation

completed
none

👉 Y-axis: Scores (math / reading / writing)

🧠 How to Read THIS Boxplot

🔹 1. Median (middle line in box)

Shows average performance

Higher line = better marks

👉 In my graph:
✔ “completed” median is higher




🔹 2. Box (middle 50% students)

Shows where most students lie

👉 Observation:
✔ Completed → scores are higher & consistent

✔ None → more spread (uneven performance)




🔹 3. Whiskers (lines)

Show overall range

👉 None group usually has:
❗ Lower minimum scores




🔹 4. Outliers (dots)

Very low or very high scores

👉 Mostly seen in:
❗ Students with no preparation




🎯 FINAL INTERPRETATION (WRITE THIS)

The boxplot shows that students who completed the test preparation course have higher median scores and more consistent performance. Students who did not complete the course have lower scores and more variation. This indicates that test preparation positively affects student performance.



## 🔗 Pairplot Analysis

Pairplot shows relationships between all numerical variables.

It helps visualize patterns and correlations between math, reading, and writing scores.


```python
sns.pairplot(df)
plt.show()
```

🧠 How to Read Pairplot


🔹 1. Diagonal graphs (middle line)

👉 These are histograms

They show:

Distribution of each subject
Example: Most students score between 60–80



🔹 2. Outside graphs (scatter plots)

👉 These show relationship between two subjects

Examples:

✔ Math vs Reading
✔ Reading vs Writing
✔ Writing vs Math

🔍 What you observe


✅ Strong relationship

👉 Reading & Writing:

Points form a straight line
Means: if reading is high → writing is also high


✅ Moderate relationship

👉 Math with Reading/Writing:

Some pattern but not perfect
Means: math is slightly different


✅ No random scatter

👉 Points are not fully scattered
→ So variables are related




🎯 FINAL INTERPRETATION (WRITE THIS)

The pairplot shows relationships between math, reading, and writing scores. Reading and writing scores have a strong positive correlation, while math shows a moderate relationship with both. The diagonal plots show the distribution of scores.
